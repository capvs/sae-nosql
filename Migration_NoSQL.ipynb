```python
Products = pandas.read_sql_query(
    """
    SELECT *
    FROM Products;
    """,
    conn
)

OrderDetails = pandas.read_sql_query(
    """
    SELECT *
    FROM OrderDetails;
    """,
    conn
)

Orders = pandas.read_sql_query(
    """
    SELECT *
    FROM Orders;
    """,
    conn
)

Customers = pandas.read_sql_query(
    """
    SELECT *
    FROM Customers;
    """,
    conn
)

Employees = pandas.read_sql_query(
    """
    SELECT *
    FROM Employees;
    """,
    conn
)

Offices = pandas.read_sql_query(
    """
    SELECT *
    FROM Offices;
    """,
    conn
)

Payments = pandas.read_sql_query(
    """
    SELECT *
    FROM Payments;
    """,
    conn
)
```

#### On regroupe Employees et Office dans Employees

```python
offices = [
    Offices.query('officeCode == @id')
        .drop(columns=["officeCode"])
        .to_dict(orient = "records")
        for id in Employees.officeCode
]

print(offices)
```

```python
Employees_offices = Employees.assign(Offices = offices)
Employees_offices.head()
```

#### On regroupe Products et OrderDetails dans OrderDetails

```python
products = [
    Products.query('productCode == @id')
        .drop(columns=["productCode"])
        .to_dict(orient = "records")
        for id in OrderDetails.productCode
]

print(products)
```

```python
details_product = OrderDetails.assign(Products = products)
details_product.head()
```

```python
#### On regroupe details_product et Orders dans Orders
```

```python
orders = [
    details_product.query('orderNumber == @id')
        .drop(columns=["orderNumber"])
        .to_dict(orient = "records")
        for id in Orders.orderNumber
]

print(orders)
```

```python
Orders_product = Orders.assign(OrderDetails_product = orders)
Orders_product.head()
```

#### On exporte les tables sur MongoDB

```python
db.Employees_offices.insert_many(
    Employees_offices.to_dict(orient = "records")
)
```

```python
db.Orders_product.insert_many(
    Orders_product.to_dict(orient = "records")
)
```

```python
db.Customers.insert_many(
    Customers.to_dict(orient = "records")
)
```

```python
db.Payments.insert_many(
    Payments.to_dict(orient = "records")
)
```
